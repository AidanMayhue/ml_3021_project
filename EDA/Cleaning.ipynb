# Cleaning the United Nations Data

### Education

```python
# Education
edu = pd.read_csv("../data/un_data/Education.csv")
edu.head()
```

```python
edu_pivot = edu.pivot_table(index = ["Country/Region Name", "Year"], columns = "Series", values = "Value", aggfunc = "first").reset_index()
edu_pivot.columns.name = None
edu_pivot.head()
```

```python
# Needs fixing
for col in edu_pivot.columns[1:]:
    print(edu_pivot[col].apply(type).unique())

# Fixing
for col in edu_pivot.columns[1:]:
    edu_pivot[col] = edu_pivot[col].astype(str).str.replace(",", "")
    edu_pivot[col] = pd.to_numeric(edu_pivot[col], errors = "coerce")
    print("\n", edu_pivot[col].apply(type).unique())
```

```python
# No NaN rows
print(len(edu_pivot[edu_pivot.isna().all(axis = 1)]))
edu_pivot.head()
```

### Merging Data into Single UN Dataframe

```python
un = edu_pivot.merge(gender_gov_pivot, on = ["Country/Region Name", "Year"], how = "outer").merge(labor_pivot, on = ["Country/Region Name", "Year"], how = "outer").merge(crime_pivot, on = ["Country/Region Name", "Year"], how = "outer").merge(water_san_pivot, on = ["Country/Region Name", "Year"], how = "outer").merge(gdp_pivot, on = ["Country/Region Name", "Year"], how = "outer").merge(pop_pivot, on = ["Country/Region Name", "Year"], how = "outer").merge(internet_pivot, on = ["Country/Region Name", "Year"], how = "outer")
un = un.rename(columns = {'Country/Region Name' : 'Country'})
un.to_csv("../data/UN_CLEAN.csv")
un.head()
```

# Cleaning the Freedom House Index Data
