```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Education
edu = pd.read_csv("../data/un_data/Education.csv")
edu = edu[edu["Country/Region Name"] != "Total, all countries or areas"]
edu.head()
```

```python
edu_pivot = edu.pivot_table(index = "Country/Region Name", columns = "Series", values = "Value", aggfunc = "first").reset_index()
edu_pivot.columns.name = None
edu_pivot.head()
```

```python
# Needs fixing
for col in edu_pivot.columns[1:]:
    print(edu_pivot[col].apply(type).unique())

# Fixing
for col in edu_pivot.columns[1:]:
    edu_pivot[col] = edu_pivot[col].astype(str).str.replace(",", "")
    edu_pivot[col] = pd.to_numeric(edu_pivot[col], errors = "coerce")
    print("\n", edu_pivot[col].apply(type).unique())


edu_pivot.head()
```
